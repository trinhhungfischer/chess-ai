# Import library and import data

```python
import chess
import chess.pgn as pgn
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import os
```

# ETL data

```python
players = ['Caruana', 'Polgar', 'Fischer', 'Morphy', 'Botvinnik', 'Tal',
 'Kasparov', 'Anand', 'Carlsen', 'Nakamura', 'Alekhine', 'Capablanca']
```

```python
white_chess_pieces = ['P', 'R', 'N', 'B', 'Q']
mapp = {'P': [0] * 400, 'N': [0] * 400, 'B': [0] * 400, 'R': [0] * 400, 'Q': [0] * 400}

for player in players:
  gameList = os.listdir('../raw_game/' + player)
  for game in gameList:
    gameLink = '../raw_game/' + player + '/' + game
    
    pgn = open(gameLink)

    first_game = chess.pgn.read_game(pgn)
    board = first_game.board()
    

    for turn, move in enumerate(first_game.mainline_moves()):
      board.push(move)
      for piece in white_chess_pieces:
        mapp[piece][turn] += str(board).count(piece)

white_data = pd.DataFrame(mapp)
```

```python
black_chess_pieces = ['p', 'r', 'n', 'b', 'q']
mapp = {'p': [0] * 400, 'n': [0] * 400, 'b': [0] * 400, 'r': [0] * 400, 'q': [0] * 400}

for player in players:
  gameList = os.listdir('../raw_game/' + player)
  for game in gameList:
    gameLink = '../raw_game/' + player + '/' + game
    
    pgn = open(gameLink)

    first_game = chess.pgn.read_game(pgn)
    board = first_game.board()
    

    for turn, move in enumerate(first_game.mainline_moves()):
      board.push(move)
      for piece in black_chess_pieces:
        mapp[piece][turn] += str(board).count(piece)

black_data = pd.DataFrame(mapp)
```

```python
piece_data = white_data.join(black_data)
piece_data.insert(loc=0, column = 'Turn', value = range(1, len(piece_data) + 1))
piece_data.to_csv('../analysis/piece-analysis.csv', index=False)
```
